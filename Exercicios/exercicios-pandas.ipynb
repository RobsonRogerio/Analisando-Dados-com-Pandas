### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
```

```python
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
```

```python
df['PREÇO MÉDIO REVENDA'] = pd.to_numeric(df['PREÇO MÉDIO REVENDA'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
df['MES'] = df['DATA FINAL'].dt.month
```

```python
df['ANO'] = df['DATA FINAL'].dt.year
```

```python
df["ANO-MES"] = df["DATA FINAL"].apply(lambda x: "{}".format(x.year)) + df["DATA FINAL"].apply(lambda x: "-{:02d}".format(x.month))
```

```python
df["ANO-MES"] = df["DATA FINAL"].apply(lambda x: f'{x.year}') + df["DATA FINAL"].apply(lambda x: f'-{x.month:02d}')
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_gcomum
```

```python
df_max = df_gcomum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].max()
df_max
```

```python
idx_min = df_gcomum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmin()
idx_min
```

```python
idx_max = df_gcomum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmax()
idx_max
```

```python
df_min = df_gcomum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].min()
df_min
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff
```

```python
df_diff['max_value'] = df_max
df_diff['min_value'] = df_min
df_diff['abs_diff'] = df_max - df_min
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
df_diff
```

```python
df_gcomum.loc[idx_max][['ESTADO', 'PREÇO MÉDIO REVENDA', 'ANO-MES']]
```

```python
df_gcomum.loc[idx_max]['ESTADO'].values
```

```python
df_diff['ESTADO MAX'] = df_gcomum.loc[idx_max]['ESTADO'].values
df_diff['ESTADO MIN'] = df_gcomum.loc[idx_min]['ESTADO'].values
df_diff
```
